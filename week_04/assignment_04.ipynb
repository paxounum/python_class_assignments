```python
# isolate just the comments dictionary
result = data['results']
comments = result['comments']
print(comments)
```

```python
# look at just one of the comments
sample = comments[3]
sample_body = sample['commentBody']
print(sample_body)
```

```python
# lowercase the comment and remove punctuation
# you can use the code we wrote for the lexical richness exercise to do this
sample_lwr = sample_body.lower()
sample = sample_lwr.replace('.','')
print(sample)
```

```python
# split text into list
word_list = sample.split(' ')
print(word_list)
```

```python
# create a dictionary of word frequencies!

# go through each word in text_data
    # if the word isn't in the dictionary, add it as a key AND give it a value of 1
    # if the word IS in the dictionary, add 1 to the value

comment_dict = {}
for text in word_list:
    if text not in comment_dict:
        comment_dict.update({text:1})
    elif text in comment_dict:
        comment_dict[text] += 1
print(comment_dict)
        
    
```

```python
# try and plot out a chart of the dictionary we just created!
import matplotlib.pyplot as plt
import numpy as np

word = comment_dict.keys()
occurrence = comment_dict.values()
print(word)
print(occurrence)
```

```python
plt.bar(range(len(comment_dict)), list(comment_dict.values()), align='center')
plt.xticks(range(len(comment_dict)), list(comment_dict.keys()), rotation = 45)
plt.show()
```
